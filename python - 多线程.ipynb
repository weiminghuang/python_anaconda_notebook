```python
import time
import threading as thread
def loop1(in1):
    # ctime 得到当前时间
    print('Start loop 1 at:',time.ctime())
    print('我是参数',in1)
    #睡眠多长时间，单位秒
    time.sleep(4)
    print('End loop 1 at:',time.ctime())
def loop2(in1,in2):
    # ctime 得到当前时间
    print('Start loop 1 at:', time.ctime())
    # 睡眠多长时间，单位秒
    print("我是",in1,"和参数",in2)
    time.sleep(2)
    print('End loop 1 at:', time.ctime())
def main():
    print("Starting at:", time.ctime())
    #启动多线程的意思是用多线程去执行某个函数
    # 启动多线程函数为 start_new_thread
    #参数两个，一个是需要运行的函数名，第二个是函数的参数作为元祖使用，为空则使用空元组
    # 注意:如果函数只有一个参数，选哟参数由一个逗号
    #生成threading.Thread实例
    t1 = thread.Thread(target=loop1,args=("王老大",))#注意，当元组只有一个参数的时候，要记得加上逗号
    t1.start()
    t2 = thread.Thread(target=loop2,args=("王大鹏","王小鹏"))
    t2.start()
    t1.join()#等待其他线程结束
    t2.join()#等待其他线程结束才执行和后面的语句
    print("All done at:",time.ctime())
if __name__ == '__main__':
    main()
    #一定要有while语句
    #因为启动多线程，本程序就作为主线程存在
    #如果主线程执行完毕，则子线程可能也需要终止
    while True:
        time.sleep(10)
```

- 守护线程-daemaon
    - 如果在程序中将子线程设置成守护线程，则子线程会在主线程结束的时候自动退出
    - 一般认为，守护线程不重要或者不允许离开主线程独立运行
    - 守护线程案例能否有效果跟环境相关
    - 案例06为非守护
- 线程常用属性
    - threading.currentThread:返回当前线程变量
    - threading.enumerate:返回一个包含正在运行的线程的list,正在运行的线程启动后，结束前的状态
    - threading.activeCount:返回正在运行的线程的数量，效果跟len(threading.enumerate)相同
    - thr.setName:给线程设置名字
    - thr.getName:得到线程的名字
- 直接继承theading.Thread

    - 重写run函数

```python
#案例06
import time
import threading
def fun():
    print("Start fun")
    time.sleep(2)
    print("end fun")
print("Main thread")
t1= threading.Thread(target=fun,args=())
t1.start()
time.sleep(1)
print("Main thread end")
```

```python
#案例7 守护线程
import time
import threading
def fun():
    print("start fun")
    time.sleep(2)
    print("end fun")
t1 = threading.Thread(target=fun, args=())
#设置守护线程的方法，必须在start (启动线程之前)之前设置，否则无效
t1.setDaemon(True)  #这行表示设置t1为守护线程  或t1.daemon = True
t1.start()
time.sleep(1)
print("main thread end")
```

```python
#案例 8
import time
import threading
def loop1():
    #ctime 得到当前时间
    print('Start loop 1 at:'.time.ctime())
    #睡眠多长时间，单位是秒
    time.sleep(4)
    print('End loop1 at:',time.ctime())

def loop2():
    # ctime 得到当前时间
    print('Start loop 2 at:'.time.ctime())
    # 睡眠多长时间，单位是秒
    time.sleep(2)
    print('End loop2 at:', time.ctime())
def loop3():
    #ctime 得到当前时间
    print('Start loop 3 at:'.time.ctime())
    #睡眠多长时间，单位是秒
    time.sleep(5)
    print('End loop3 at:',time.ctime())
def main():
    print("Start at :",time.ctime())
    #生成threading.Thread实例
    t1 = threading.Thread(target=loop1, args=())
    t1.setName("THR_1")
    t1.start()
    t2 = threading.Thread(target=loop2, args=())
    t2.setName("THR_2")
    t2.start()
    t3 = threading.Thread(target=loop3, args=())
    t3.setName("THR_3")
    t3.start()
    #预期3秒后，thread2已经自动结束
    time.sleep(3)
    # enumerate 得到正在运行子线程，即子线程1和子线程3
    for thr in threading.enumerate():
        # getName能够得到线程的名称
        print("正在运行的线程名字是:{0}".format(thr.getName()))
    print("正在运行的子线程数量wei:{0}".format(threading.activeCount()))
    print("All done at:",time.ctime())
if __name__ =="__main__":
    main()
    #一定要有while语句
    #因为启动多线程，本程序就作为主线程存在
    #如果主线程执行完毕，则子线程可能也需要终止
    while True:
        time.sleep(10)
```

```python
#案例9
import threading
import time
#1.类需要继承自threading.Thead
class MyThread(threading.Thread):
    def __init__(self,arg):
        super(MyThread,self).__init__()# 一定要写这个
        self.arg = arg
        #2 必须重写run函数，run函数代表的是真正执行的功能
    def run(self):
        time.sleep(2)
        print("The args for this class is {0}".format(self.arg))
for i in range(5):
    t = MyThread(i):
    t.start()
    t.join()
print("Main thread is done !!!!!!")
```
