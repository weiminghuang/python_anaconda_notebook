# 多线程vs多进程
- 程序：一堆代码以文本形式存人文档
- 进程：程序运行的一个状态
    - 包含地址控件，内存，数据栈等
    - 每个进程由自己完全独立的运行环境，多进程共享数据是一个问题
- 线程：
    - 一个进程的独立运行片段
    - 轻量化的进程
    - 一个进程的多个线程间共享数据和上下问运行环境
    - 共享互斥问题
- 全局解释器锁（GIL）
    - Python代码的执行是由python虚拟机进行控制
    - 在主循环中只能有一个控制线程在执行

- Python包
    - thread:有问题，不好用，python3改成了_thread
    - threading：通信的包

```python
#案例12
import threading
sum = 0
loopSum = 1000000
lock = threading.lock()  # 生成锁实例
def myAdd():
    global sum ,loopSum
    for i in range(1,loopSum):
        #上锁，申请锁
        lock.acquire()
        sum += 1
        # 释放锁   注:上锁后，不用了，一定要记得解锁
        lock.release()
def myMinu():
    global sum,loopSum
    for i in range(1,loopSum):
          #上锁，申请锁
        lock.acquire()
        sum -= 1
         # 释放锁   注:上锁后，不用了，一定要记得解锁
        lock.release()
if __name__ == '__main__':
    print("Starting .....{0}".format(sum))
    #开启多线程的实例，看执行结果是否一样
    t1 = threading.Thread(target=myAdd, args=())
    t2 = threading.Thread(target=myMinu,args=())
    t1.start()
    t1.join()
    t2.join()
    print("Done ...{0}".format(sum))
```

```python
# 案例20
import multiprocessing
from time import sleep,ctime
class ClockProcess(multiprocessing.Process):
    '''
     两个函数比较重要
     1 init构造函数
     2 run
    '''
    def __init__(self,interval):
        super().__init__()
        self.interval = interval
    def run(self):
        while True:
            print("the time is %s" % ctime())
            sleep(self.interval)
if __name__ == '__main__'
    p = ClockProcess(3)
    p.start()
    while True:
        print('sleeping......')
        sleep(1)
```

```python
#案例23
import multiprogressing
from time import ctime
def consumer(input_q):
    print("Into consumer:".ctime())
    while True:
        #处置项
        item = input_q.get()
        if item is None:
            break
        print("pull",item,"out of q")
    print("Out of consumer:",ctime()) ##此处执行完成，在转入主进程
def producer(sqquence,output_q):
    print("Into procuder:",ctime())
    for item in sequence:
        output_q.put(item)
        print("put",item,'into q')
    print("Out of procuder:",ctime())
#建立进程
if __name__ == '__main__':
    q = multiprocessing.Queue()
    #允许消费者进程
    cons_p = multiprocessing.Process(target = consumer,args = (q,))
    cons_p.start()
    #生产多个项，sqquence代表要送给消费者的顺序列
    #在实践中，这可能是生成器的输出或通过一些其它方式生产出来
    sequence = [1,2,3,4]
    producer(sequence,q)
    q.put(None)
    #等待所有项被处理
    q.join()
    cons_p.join()
```
